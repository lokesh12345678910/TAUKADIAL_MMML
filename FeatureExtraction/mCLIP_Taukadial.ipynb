```python
def multiligualClipSimScore(clipModelName, textModelName, texts,imagePath="CatRescuePackage.png"):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    if clipModelName != 'ViT-B-16-plus-240':
      clip_model, preprocess = clip.load(clipModelName, device)
    else:
      import open_clip
      clip_model, _, preprocess = open_clip.create_model_and_transforms('ViT-B-16-plus-240', pretrained="laion400m_e32",device=device)
# Load the image and text
    image = preprocess(Image.open(imagePath)).unsqueeze(0).to(device)
    image_features = clip_model.encode_image(image).to(torch.float32).to(device)
    #print("Dim(Image):", image_features.shape)
    # Load Model & Tokenizer
    textEncoder = pt_multilingual_clip.MultilingualCLIP.from_pretrained(textModelName)
    tokenizer = transformers.AutoTokenizer.from_pretrained(textModelName)
    simScores = []
    for transcription in texts:
      with torch.no_grad():
        text_features = textEncoder.forward(transcription, tokenizer)
        text_features = text_features.to(torch.float32).to(device)
        final_similarity_score = (text_features @ image_features.T).mean().item()
       #print(f"Similarity Score: {final_similarity_score.item()}")
        #print("# of characters:", len(transcription))
        simScores.append(final_similarity_score)
    return simScores
```

```python
modelPairs = [('ViT-B-16-plus-240', 'M-CLIP/XLM-Roberta-Large-Vit-B-16Plus'), ('ViT-B/32', 'M-CLIP/XLM-Roberta-Large-Vit-B-32'), ('ViT-L/14', 'M-CLIP/LABSE-Vit-L-14'), ('ViT-L/14', "M-CLIP/XLM-Roberta-Large-Vit-L-14")]
```

```python
from PIL import Image

def split_into_thirds(image_path):
    # Open the image
    img = Image.open(image_path)

    # Get the width and height of the image
    width, height = img.size

    # Calculate the width of each third
    third_width = width // 3

    # Split the imamge into thirds
    first_third = img.crop((0, 0, third_width, height))
    second_third = img.crop((third_width, 0, 2 * third_width, height))
    third_third = img.crop((2 * third_width, 0, width, height))

    # Return the three thirds as PIL Image objects
    return first_third, second_third, third_third

# Example usage
image_path = "pictures-cn.png"
first_third, second_third, third_third = split_into_thirds(image_path)

# Save or display the thirds as desired
first_third.show()
second_third.show()
third_third.show()
```

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import os, random
import torch

import warnings
warnings.filterwarnings("ignore")

!pip install git+https://github.com/openai/CLIP.git
import clip
```

```python
!pip install open_clip_torch
import open_clip
```

```python
def multiligualClipSimScore(clipModelName, textModelName, texts,img):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    if clipModelName != 'ViT-B-16-plus-240':
      clip_model, preprocess = clip.load(clipModelName, device)
    else:
      import open_clip
      clip_model, _, preprocess = open_clip.create_model_and_transforms('ViT-B-16-plus-240', pretrained="laion400m_e32",device=device)
# Load the image and text
    image = preprocess(img).unsqueeze(0).to(device)
    image_features = clip_model.encode_image(image).to(torch.float32).to(device)
    #print("Dim(Image):", image_features.shape)
    # Load Model & Tokenizer
    textEncoder = pt_multilingual_clip.MultilingualCLIP.from_pretrained(textModelName)
    tokenizer = transformers.AutoTokenizer.from_pretrained(textModelName)
    simScores = []
    for transcription in texts:
      with torch.no_grad():
        text_features = textEncoder.forward(transcription, tokenizer)
        text_features = text_features.to(torch.float32).to(device)
        final_similarity_score = (text_features @ image_features.T).mean().item()
       #print(f"Similarity Score: {final_similarity_score.item()}")
        #print("# of characters:", len(transcription))
        simScores.append(final_similarity_score)
    return simScores
```

```python
modelPairs = [('ViT-B-16-plus-240', 'M-CLIP/XLM-Roberta-Large-Vit-B-16Plus'), ('ViT-B/32', 'M-CLIP/XLM-Roberta-Large-Vit-B-32'), ('ViT-L/14', 'M-CLIP/LABSE-Vit-L-14'), ('ViT-L/14', "M-CLIP/XLM-Roberta-Large-Vit-L-14")]
```

```python
firstImageMultiligualClipSimScores = []
for imageEncoder, textEncoder in modelPairs:
  print(imageEncoder, textEncoder)
  truncatedfirstImageTranscriptions = [text[:max_sequence_length] for text in firstImageTranscriptions]
  pairSimScores = multiligualClipSimScore(imageEncoder, textEncoder, truncatedfirstImageTranscriptions,firstImage)
  firstImageMultiligualClipSimScores.append(pairSimScores)
```
